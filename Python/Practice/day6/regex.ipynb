```python
# 1. match pattern startimport pandas as pd
text = "In 2025, Alex_T went to Room-42 at 10:30 AM to check sensor-data, but he noticed something strange! The device ID was 'A9B-77?' and the system displayed errors like E1, E12, and E123. Later, he wrote notes such as: temp=29C, humidity=70%, and status=OK. However, after 5 minutes, the sensor stopped responding... again! .error " 

# match pattern
if re.match(r"2025",text) == None:
    print("Not fund")
else:
    print(re.match(r"In",text).span())  

# search pattern --> similar can get span and all
if not re.search(r'2025', text):
    print("Not found")
else:
    print("Found")

# findall pattern --> list of all patterns found
re.findall(r"he",text)
```

```python
# 2. splitting based on delimeter --> in case single char and conflicting with predefined things then use \ as escapse sequence
re.split(r'[, .]',text)
```

```python
# 3. validate valid vin 
vin = "WMI12345678901237"
pattern = r"^\w{3}\d{14}$"
# pattern = r"^[A-Za-z]{3}\d{14}$"
print(re.match(pattern, vin))
```

```python
# 4. valid no plate or not 
license_plate = "MH 15 AZ 8906"
pattern = r"^[A-Za-z]{2}\s\d{2}\s[A-Za-z]{2}\s\d{4}$"
re.match(pattern,license_plate)
```

```python
# 5. extract no plate from paragraph
para = "During the routine inspection at the logistics hub, the supervisor noticed that one of the delivery vans had arrived earlier than scheduled. The driver, Rahul Verma, presented his identification documents along with his vehicle registration papers. His license number MH14 DR 7296 and MH 15 DR 8906 was recorded in the daily logbook before he proceeded to the loading zone to collect the sealed shipment boxes. Everything appeared in order, and the team allowed him to continue without delay."
pattern = r"[A-Za-z]{2}\s*\d{2}\s*[A-Za-z]{2}\s*\d{4}"
re.findall(pattern,para)
```

```python
# 6. finding "P" followed by 4 digits
text = "P0300, P0301, P0302, P0300, P34, p567"
pattern = r'P\d{4}'
re.findall(pattern, text)
```

```python
# 7. validate date
date_string = "2023-10-27"
pattern = r'^\d{4}-\d{2}-\d{2}$'
re.match(pattern, date_string)
```

```python
# 8. extracting emails --> non capturing gropu does not create group and onlu written for 'or' or 'and' logic 
log_content = "Report sent to user@example.in and support@kpit.com. support@kpit.gov"
pattern = r"\w+@\w+\.(?:com|in)"
re.findall(pattern, log_content)
```

```python
# 9. extracting speed and rpm 
can_message = "Speed: 64.5 km/h, RPM: 2500"
pattern = r"Speed:\s*\d+(?:\.\d+)?\s*km/h" # same can be done with rpm
re.findall(pattern, can_message)
```

```python
# 10. extracting key value pairs
data_string = "EngineTemp=90.5,OilPressure=50,VehicleSpeed=80.2,SensorTemp=.56"
pattern = "\w+=\d*(?:\.\d+)?"
re.findall(pattern, data_string)
```

```python
# 11. example for substitution
text = "This   string  has   too  many   spaces."
pattern = r"\s+"
re.sub(pattern," ",text)
```

```python
# 12 substitute with # for every word which ends with digit --> concept of word boundary
def replace_ending_digit_words(text):
    if not text:
        return ""
    pattern = r"\b\w*\d\b"
    return re.sub(pattern,"#",text)
    
text = """User123 logged in at 12/05/2023.\nEmail: alice@example.com\nOrder #A00123"""
replace_ending_digit_words(text)
```

```python
# 13. Find words â‰¥ 5 characters
from typing import List, Dict,Any
def words_greater_than_len5(text:str) -> List[str]:
    # pattern = r"\b\w*\b"
    # words = re.findall(pattern, text)
    # return list(filter(lambda x: len(x)>5, words))
    pattern = r"\b\w{5,}\b"
    return re.findall(pattern, text)
words_greater_than_len5(text)
```

```python
# 14. count_words_greater_than_3_digits
def words_with_more_than_3_digits(text:str)->List[str]:
    pattern = r"\b\w*(?:\d\w*){3,}\b"
    return re.findall(pattern, text)
words_with_more_than_3_digits(text)
```
