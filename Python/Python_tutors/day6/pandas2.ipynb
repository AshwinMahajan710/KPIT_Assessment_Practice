```python
import pandas as pd
```

```python
# creating dataframe
sensors = [
    {"id": 101, "timestamps": 10.5, "status": "OK",     "Temperature": 98},
    {"id": 102, "timestamps": 11.0, "status": "FAIL",   "Temperature": 105},
    {"id": 103, "timestamps": 10.8, "status": "OK",     "Temperature": 99},
    {"id": 104, "timestamps": 12.1, "status": "WARN",   "Temperature": 102},
    {"id": 105, "timestamps": 10.2, "status": "OK",     "Temperature": 97},
    {"id": 106, "timestamps": 11.4, "status": "FAIL",   "Temperature": 110},
    {"id": 107, "timestamps": 10.9, "status": "OK",     "Temperature": 100}
]

df = pd.DataFrame(sensors)
df
```

```python
# Normal funcs
df.head()
df.tail()
df.describe()
df.info()
```

```python
# accessing via loc/iloc
df.loc[0,"id"] # --> way to write is [row, col]
df.iloc[0:2,0:4]
```

```python
# Group by with same operations --> some more operations available (mean, median , nunique, min, max, count, first, last)
df.groupby("status")["Temperature"].agg(sum)

df.groupby("status").agg(
    avg_temperature = ("Temperature", "mean"),
    timestamps = ("timestamps", "mean")
)
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Tom', 'John', 'Anna', 'Peter'],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Tokyo', 'New York', 'Paris', 'Berlin'],
    'Sales': [100, 200, 300, 400, 500, 100, 200, 300]
}
df = pd.DataFrame(data)
```

```python
# pivot table
df.pivot_table(index = 'City', columns = 'Name', values = "Sales", aggfunc = 'mean')
```

```python
# sortng values with respect to any numeric columns
df.sort_values(['Sales'])
```
