```python
# ex1. validating vin valid or not
vin = "WMI12345678901237"
pattern = r"^\w{3}\d{14}&"
# pattern = r"^[A-Z0-9]{17}&" --> another way
re.match(pattern,vin)
```

```python
# ex2. valid no plate or not 
license_plate = "MH 15 AZ 8906"
pattern = r"^\w{2}\s*\d{2}\s*\w{2}\s*\d{4}\s*"
re.match(pattern, license_plate)
```

```python
# ex3. apply in para
para = "During the routine inspection at the logistics hub, the supervisor noticed that one of the delivery vans had arrived earlier than scheduled. The driver, Rahul Verma, presented his identification documents along with his vehicle registration papers. His license number MH14 DR 7296 and MH 15 DR 8906 was recorded in the daily logbook before he proceeded to the loading zone to collect the sealed shipment boxes. Everything appeared in order, and the team allowed him to continue without delay."
pattern =  r"\w{2}\s*\d{2}\s*\w{2}\s*\d{4}\s*"
re.findall(pattern, para)
```

```python
# ex4. finding "P" followed by 4 digits
text = "P0300, P0301, P0302, P0300, P34, p567"
pattern = "P\d{4}"
re.findall(pattern, text)
```

```python
# ex5. validate date
date_string = "2023-10-27"
pattern = "\d{4}-\d{2}-\d{2}"
re.findall(pattern, date_string)
```

```python
import re

#  ex6. extracting email --> just remember how the captured group is written 
log_content = "Report sent to user@example.in and support@kpit.com . support@kpit.gov"
pattern = r"\w+@\w+\.(?:com|in)?"
re.findall(pattern, log_content)
```

```python
# ex.7 extracting speed and rpm
can_message = "Speed: 65.2 km/h, RPM: 2500"
pattern_speed = r"Speed:\s*\d+(?:\.\d+)?\s*km/h"
pattern_rpm = "RPM:\s*\d{4}"
pattern_without_KMH = r"Speed:\s*\d+(?:\.\d+)?"
pattern_with_grps = r"(Speed:\s*\d+(?:\.\d+)?\s*km/h), (RPM:\s*\d+)"
print(re.findall(pattern_speed, can_message))
print(re.findall(pattern_rpm, can_message))
print(re.findall(pattern_without_KMH, can_message))
print(re.findall(pattern_with_grps, can_message)[0][0])
print(re.findall(pattern_with_grps, can_message)[0][1])
```

```python
# ex8. extracting key value pairs
data_string = "EngineTemp=90.5,OilPressure=50,VehicleSpeed=80.2,SensorTemp=.56"
pattern = r"([A-Za-z]+)=(\d+(?:\.\d+)?|\.\d+)"
print(re.findall(pattern, data_string))
```

```python
# ex.9 example for substitution
text = "This   string  has   too  many   spaces."
pattern = r"\s+"
cleaned_text = re.sub(pattern, " ", text)
print("Cleaned Text:", cleaned_text)
```

```python
 # ex.10 substitute with # for every word which ends with digit --> concept of word boundary
def replace_ending_digit_words(text: str) -> str:
    if not text:
        return ""
    return re.sub(r"\b\w*\d\b", '#',text)

text = """User123 logged in at 12/05/2023.\nEmail: alice@example.com\nOrder #A00123"""
replace_ending_digit_words(text)
```
