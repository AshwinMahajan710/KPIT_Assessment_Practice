```python
# 1. creating the dataframe
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Tom','eshan','aman','ram'],
    'Age': [28, 24, 35, 32, 40,33,44,55],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Tokyo','city1','city2','city3']
}

df = pd.DataFrame(data)
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'Tom', 'John', 'Anna', 'Peter'],
    'City': ['New York', 'Paris', 'Berlin', 'London', 'Tokyo', 'New York', 'Paris', 'Berlin'],
    'Sales': [100, 200, 300, 400, 500, 100, 200, 300]
}
df = pd.DataFrame(data)
```

```python
# 3. group-by ops --> grouping with which attribute and use one aggregator
df.groupby('Name').sum()
```

```python
# also can extract particular columns if want to
df.groupby('Name')["Sales"].sum()
```

```python
# 4. also can create the pivot_tables --> 
df.pivot_table(index='Name',columns = 'City',values='Sales',aggfunc='sum')
```

```python
# 5. Joins and merge
# Create sample DataFrames
data1 = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32]
}
df1 = pd.DataFrame(data1)

data2 = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df2 = pd.DataFrame(data2)
```

```python
# Join operation
print(df1.set_index('Name').join(df2.set_index('Name'))) # --> using 'join' seems difficult
newdf = pd.merge(df1,df2,on="Name") # --> using merge it is easy
newdf
```

```python
# 6. for dropping NA --> df.dropna()
# 7. for dropping duplicates -> df.drop_duplicates()
```

```python
# 8. Sorting with respect to age
newdf.sort_values("Age") 
```

```python
# # 9. Group by status and calculate count and average temperature 
# Define the SensorHealthReport data
# if want column with personalize name and agg type use following way
sensor_data = [
    {"sensor_id": "TEMP_01", "timestamp": 10.5, "status": "OK", "temperature": 45.2},
    {"sensor_id": "BRAKE_02", "timestamp": 20.1, "status": "WARN", "temperature": 60.3},
    {"sensor_id": "TEMP_01", "timestamp": 30.7, "status": "FAIL", "temperature": 75.0},
    {"sensor_id": "BRAKE_02", "timestamp": 40.2, "status": "OK", "temperature": 50.1},
    {"sensor_id": "TEMP_01", "timestamp": 50.9, "status": "WARN", "temperature": 65.4}
]
# Convert to DataFrame
df = pd.DataFrame(sensor_data)
df.groupby("status").agg(
    count = ("status", "size"),
    average_temperature=("temperature", "mean")
)
```
