```python
# opening the file --> some basic operation
# error if --> file not exist and try to open it using the 'x' mode
# read() --> to read the content in file, write() --> to write content in file
# seek(0) --> move the pointer at the start of the file so that we can read it
# w+ --> read + write , r+ --> only read , a --> for appending info, a+ --> for append with read op
# close the file after use
# even for truncate just open file in 'w+' mode
file = open('Hello.txt','w+')
file.read()
file.write("Hello ashwin.\nToday is your first day for corporate job, isn't it?")
file.seek(0)
file.read()
file.close()

# to modify must use a+ so that we can both modify and append the changes
# --> There is no method called as 'append' .. use 'write' only

file = open('Hello.txt','a+')
file.write("\nHope you are getting a new message")
file.seek(0)
file.read()
file.close()

# using list we can write in the file --> writelines
# do not ever forgot to set the pointer again
# also can get the input line by line
# also can read the line using 'readline' only --> and can use tell() to get the pointer location
# also for seek with particular offset() --> # f.seek(6,0) # (offset-howmanybytestomove,fromwheretostart)
file = open('Hello.txt','a+')
list1 = ["Ashwin The best","\nAshwin the great","\nAshwin the phenomenal"]
file.writelines(list1)
# file.seek(0)
file.readlines()
file.close()

# os methods --> os.getcwd(), os.rename(), os.remove()
```
